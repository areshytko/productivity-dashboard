## Load data

```python
from dashboard.load import load
import dashboard.config as config 
from dashboard.gheets import ManualFlow

import json
```

```python
flow = ManualFlow()
url = flow.get_url()
with open('notebooks/url.json', 'w') as wf:
    json.dump(url, wf)
```

```python
creds = flow.get_google_token()
```

```python
from dashboard.load import load_pomodoros, load_catalog
pomodoros = load_pomodoros(
        credentials=creds,
        spreadsheet_id=config.POMODOROS_SPREADSHEET_ID,
        range_name=config.POMODOROS_RANGE,
    )
catalog = load_catalog(
    credentials=creds,
    spreadsheet_id=config.ACTIVITIES_SPREADSHEET_ID,
    range_name=config.ACTIVITIES_RANGE
)
```

```python
from dashboard.load import fill_all_activities

res = fill_all_activities(pomodoros, catalog)
```
